```python
plt.figure(figsize=(15, 15))

for i in range(0, len(numeric_features)):
    plt.subplot(5, 3, i+1)
    sns.histplot(x=df[numeric_features[i]], color='r')  #to understand the distribution of a dataset.
    plt.xlabel(numeric_features[i])
    plt.tight_layout()
# tight_layout automatically adjusts subplot parameters to give better spacing between plots and labels. 
# It helps prevent overlapping elements like axis labels, titles.
```

```python
def detect_outliers(col):
    # Finding the IQR
    percentile25 = df[col].quantile(0.25)
    percentile75 = df[col].quantile(0.75)
    print('\n', col )
    print("percentile25",percentile25)
    print("percentile75",percentile75)
    iqr = percentile75 - percentile25
    upper_limit = percentile75 + 1.5 * iqr
    lower_limit = percentile25 - 1.5 * iqr
    print("Upper limit",upper_limit)
    print("Lower limit",lower_limit)
    df.loc[(df[col]>upper_limit), col]= upper_limit
    df.loc[(df[col]<lower_limit), col]= lower_limit    
    return df
```

**Why are we capping it and why not trim it ?**
* Trimming outliers may result in the removal of a large number of records from this dataset as we have already very less rows so this isn’t desirable in this case since columns other than the ones containing the outlier values may contain useful information.

* In this cases, you can use outlier capping to replace the outlier values with a maximum or minimum capped values. Be warned, this manipulates our data but we can replace outlier values by the upper and lower limit calculated using the IQR range.

```python
for col in continuous_features:
         detect_outliers(col)
```

```python
all_models_data = []  # This will hold dictionaries, which will become rows
def mymodel(model,model_name):
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    mae=mean_absolute_error(y_test,y_pred)
    mse=mean_squared_error(y_test,y_pred)
    rmse=np.sqrt(mse)
    r2=r2_score(y_test,y_pred)
    print(f'MAE: {mae} \nMSE: {mse} \nRMSE: {rmse} \n R2 Score: {r2}')
    train_score= model.score(X_train,y_train)
    test_score= model.score(X_test,y_test)
    
    print(f'Training Performance : {train_score}')
    print(f'Testing Performance : {test_score}')
    if model_name != " ":
        model_data = model_store(model_name, r2, train_score-test_score)  # Get the dictionary

        global all_models_data  # Access the global list
        all_models_data.append(model_data)  # Add the dictionary to the list

        combined_df = pd.DataFrame(all_models_data)
    return model
```

```python
def model_store(model_name,r2_score,Error):
    model_data = {
        "model_name": model_name,
        "r2_score": r2_score,
        "Error": Error
    }
    return model_data
```

```python
def predict_price(model, scaler, df_train):
    print("Enter the car details:")
    brand_name = str(input("Brand Name: "))
    vehicle_age = eval(input("Vehicle Age: "))
    km_driven = eval(input("Kilometers Driven: "))
    seller_type = input("Seller Type (Individual/Dealer): ")
    fuel_type = input("Fuel Type (Petrol/Diesel/CNG/LPG): ")
    transmission_type = input("Transmission Type (Manual/Automatic): ")
    mileage = float(input("Mileage (kmpl): "))
    engine = eval(input("Engine Capacity (cc): "))
    max_power = float(input("Max Power (bhp): "))
    seats = eval(input("Number of Seats: "))

    input_dict = {
        "brand":brand_name,
        "vehicle_age": vehicle_age,
        "km_driven": km_driven,
        "seller_type": seller_type,
        "fuel_type": fuel_type,
        "transmission_type": transmission_type,
        "mileage": mileage,
        "engine": engine,
        "max_power": max_power,
        "seats": seats
    }

    input_df = pd.DataFrame([input_dict])
    input_df = pd.get_dummies(input_df, columns=["brand","seller_type", "fuel_type", "transmission_type"],dtype=int)   

    missing_cols = set(df_train.drop(columns=["log_price"]).columns) - set(input_df.columns)
    for col in missing_cols:
        input_df[col] = 0
    input_df = input_df[df_train.drop(columns=["log_price"]).columns]
    input_data = scaler.transform(input_df)
    # print(input_data)
    predicted_price = model.predict(input_data)
    # Exponentiate to get the predicted selling price
    predicted_price = np.exp(predicted_price)
    print(f"Predicted Selling Price: ₹{predicted_price[0]}")
```
